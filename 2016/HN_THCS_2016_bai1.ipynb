Cho mt dy s nguyn dng gm n phn t (1n100). Vit chng trnh thc hin cc yu cu sau:
Nhp n v cc phn t ca dy s.
Tm gi tr ln th nh ca dy s.
D liu vo:  c cho trong tp vn bn c tn dayso.inp gm: Dng u cha s nguyn n; Dng tip theo, cha n s nguyn dng, cc s cch nhau t nht mt du cch.
Kt qu ra: c ghi trong mt tp vn bn c tn dayso.out gm mt dng cha s ln th nh trong dy. Nu khng tm c s nh vy th in ra duy nht gi tr -1


```python
def read_input(file_name):
    with open(file_name, 'r') as file:
        n = int(file.readline().strip())
        arr = list(map(int, file.readline().strip().split()))
    return n, arr

def find_second_largest(arr):
    first = second = -1
    for number in arr:
        if number > first:
            second = first
            first = number
        elif number > second and number != first:
            second = number
    return second

def write_output(file_name, result):
    with open(file_name, 'w') as file:
        file.write(str(result))

if __name__ == "__main__":
    n, arr = read_input('basic_python/2016/dayso.inp')
    if n < 2:
        result = -1
    else:
        result = find_second_largest(arr)
    write_output('basic_python/2016/dayso.out', result)
```

read_input(file_name): Hm ny c d liu t tp dayso.inp. Dng u tin cha s nguyn n. Dng th hai cha n s nguyn dng.

find_second_largest(arr): Hm ny tm gi tr ln th nh trong dy s. Nu khng tm c, tr v -1.

write_output(file_name, result): Hm ny ghi kt qu vo tp dayso.out.
