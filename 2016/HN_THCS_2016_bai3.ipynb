```python
def find_generators(M):
    generators = []
    for N in range(1, M):
        digit_sum = sum(int(digit) for digit in str(N))
        if N + digit_sum == M:
            generators.append(N)
    return generators

# Test
M = 216
M = 4501010

generators = find_generators(M)
if generators:
    print(f"Tất cả các số N có thể sinh ra {M} là: {generators}")
    print(f"Số nhỏ nhất sinh ra {M} là: {min(generators)}")
else:
    print(f"Không có số nào sinh ra {M}")

```
