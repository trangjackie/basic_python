# Thử nghiệm robot
Công ty HP vừa thiết kế một loại robot thông minh mới. Để đánh giá khả năng tự vận hành của robot, người ta tạo ra một bức tường từ n ݊cột các khối lập phương, các cột đặt cạnh nhau, bề dày bức tường là 1, độ cao cột thứ i là ܽa_i (do ܽa_i khối lập phương tạo lên). Có ݉m robot tham gia thử nghiệm. Trước tiên người ta chia n cột thành m đoạn bằng ݉m − 1 điểm cắt k_1, k_2, ..., k_(m-1) (݇ k_0 = 0 < k_1 < .. < k_(m-1) < k_m = n). Robot thứ i được giao nhiệm vụ xếp lại đoạn từ cột k_(i-1) + 1 đến cột k_i sao cho các cột trong đoạn có độ cao bằng nhau. Robot chỉ có thể thực hiện một trong hai loại thao tác, mỗi thao tác mất 1 đơn vị thời gian. 
• Thao tác 1: Lấy khối trên cùng của một cột trong đoạn được giao để bỏ đi; 
• Thao tác 2: Lấy một khối mới, đặt khối đó lên trên cùng của một cột trong đoạn được giao.

Thời gian kết thúc thử nghiệm là thời gian mà robot cuối cùng hoàn thành xong nhiệm vụ. 
Yêu cầu: Cho a_1, a_2, ..., a_n và m. Hãy tìm m − 1 điểm cắt để chia n cột thành m đoạn sao cho thời gian thử nghiệm là nhanh nhất, biết các robot đều thực hiện các thao tác tối ưu.

Dữ liệu: Vào từ thiết bị nhập chuẩn: 
- Dòng đầu chứa hai số nguyên n và m;
- Dòng thứ hai gồm n số nguyên không âm a_1, a_2, .., a_n ( a_i ≤ 10^6).
Kết quả: Ghi ra thiết bị ra chuẩn một dòng chứa một số nguyên là thời gian ít nhất để thử nghiệm.

```python
def min_time_to_balance_columns(n, m, a):
    # Tính tổng các đoạn
    prefix_sum = [0] * (n + 1)
    for i in range(n):
        prefix_sum[i + 1] = prefix_sum[i] + a[i]

    # Tính tổng bình phương các đoạn
    prefix_square_sum = [0] * (n + 1)
    for i in range(n):
        prefix_square_sum[i + 1] = prefix_square_sum[i] + a[i] * a[i]

    # Hàm tính tổng thời gian cần thiết cho đoạn từ cột i đến cột j
    def calc_time(i, j):
        total = prefix_sum[j + 1] - prefix_sum[i]
        mean = total / (j - i + 1)
        return int(prefix_square_sum[j + 1] - prefix_square_sum[i] - mean * total)

    # Bảng dp lưu trữ thời gian tối thiểu
    dp = [[float('inf')] * n for _ in range(m)]
    for i in range(n):
        dp[0][i] = calc_time(0, i)

    # Tính dp cho các robot
    for k in range(1, m):
        for i in range(k, n):
            for j in range(k - 1, i):
                dp[k][i] = min(dp[k][i], dp[k - 1][j] + calc_time(j + 1, i))

    return dp[m - 1][n - 1]

# Đọc dữ liệu vào
n, m = map(int, input().split())
a = list(map(int, input().split()))

# Tính toán và in ra kết quả
result = min_time_to_balance_columns(n, m, a)
print(result)
```

```python
def min_time_to_balance_columns(n, m, a):
    prefix_sum = [0] * (n + 1)
    for i in range(n):
        prefix_sum[i + 1] = prefix_sum[i] + a[i]

    def calc_time(i, j):
        total = prefix_sum[j + 1] - prefix_sum[i]
        mean = total // (j - i + 1)
        time = 0
        for k in range(i, j + 1):
            time += abs(a[k] - mean)
        return time

    dp = [[float('inf')] * (n + 1) for _ in range(m + 1)]
    for i in range(n + 1):
        dp[0][i] = 0

    for k in range(1, m + 1):
        for i in range(k, n + 1):
            for j in range(k - 1, i):
                dp[k][i] = min(dp[k][i], dp[k - 1][j] + calc_time(j, i - 1))

    return dp[m][n]

# Đọc dữ liệu vào
n, m = map(int, input().split())
a = list(map(int, input().split()))

# Tính toán và in ra kết quả
result = min_time_to_balance_columns(n, m, a)
print(result)
```

```python
def min_time_to_balance_columns(n, m, a):
    prefix_sum = [0] * (n + 1)
    for i in range(n):
        prefix_sum[i + 1] = prefix_sum[i] + a[i]

    def calc_time(i, j):
        total = prefix_sum[j + 1] - prefix_sum[i]
        mean = total // (j - i + 1)
        time = sum(abs(a[k] - mean) for k in range(i, j + 1))
        return time

    dp = [[float('inf')] * (n + 1) for _ in range(m + 1)]
    dp[0][0] = 0

    for k in range(1, m + 1):
        for i in range(k, n + 1):
            for j in range(k - 1, i):
                dp[k][i] = min(dp[k][i], dp[k - 1][j] + calc_time(j, i - 1))

    return dp[m][n]

# Đọc dữ liệu vào
n, m = map(int, input().split())
a = list(map(int, input().split()))

# Tính toán và in ra kết quả
result = min_time_to_balance_columns(n, m, a)
print(result)
```
