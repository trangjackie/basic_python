# Th nghim robot
Cng ty HP va thit k mt loi robot thng minh mi.  nh gi kh nng t vn hnh ca robot, ngi ta to ra mt bc tng t n ct cc khi lp phng, cc ct t cnh nhau, b dy bc tng l 1,  cao ct th i l a_i (do a_i khi lp phng to ln). C m robot tham gia th nghim. Trc tin ngi ta chia n ct thnh m on bng m  1 im ct k_1, k_2, ..., k_(m-1) ( k_0 = 0 < k_1 < .. < k_(m-1) < k_m = n). Robot th i c giao nhim v xp li on t ct k_(i-1) + 1 n ct k_i sao cho cc ct trong on c  cao bng nhau. Robot ch c th thc hin mt trong hai loi thao tc, mi thao tc mt 1 n v thi gian. 
 Thao tc 1: Ly khi trn cng ca mt ct trong on c giao  b i; 
 Thao tc 2: Ly mt khi mi, t khi  ln trn cng ca mt ct trong on c giao.

Thi gian kt thc th nghim l thi gian m robot cui cng hon thnh xong nhim v. 
Yu cu: Cho a_1, a_2, ..., a_n v m. Hy tm m  1 im ct  chia n ct thnh m on sao cho thi gian th nghim l nhanh nht, bit cc robot u thc hin cc thao tc ti u.

D liu: Vo t thit b nhp chun: 
- Dng u cha hai s nguyn n v m;
- Dng th hai gm n s nguyn khng m a_1, a_2, .., a_n ( a_i  10^6).
Kt qu: Ghi ra thit b ra chun mt dng cha mt s nguyn l thi gian t nht  th nghim.

```python
def min_time_to_balance_columns(n, m, a):
    # Tnh tng cc on
    prefix_sum = [0] * (n + 1)
    for i in range(n):
        prefix_sum[i + 1] = prefix_sum[i] + a[i]

    # Tnh tng bnh phng cc on
    prefix_square_sum = [0] * (n + 1)
    for i in range(n):
        prefix_square_sum[i + 1] = prefix_square_sum[i] + a[i] * a[i]

    # Hm tnh tng thi gian cn thit cho on t ct i n ct j
    def calc_time(i, j):
        total = prefix_sum[j + 1] - prefix_sum[i]
        mean = total / (j - i + 1)
        return int(prefix_square_sum[j + 1] - prefix_square_sum[i] - mean * total)

    # Bng dp lu tr thi gian ti thiu
    dp = [[float('inf')] * n for _ in range(m)]
    for i in range(n):
        dp[0][i] = calc_time(0, i)

    # Tnh dp cho cc robot
    for k in range(1, m):
        for i in range(k, n):
            for j in range(k - 1, i):
                dp[k][i] = min(dp[k][i], dp[k - 1][j] + calc_time(j + 1, i))

    return dp[m - 1][n - 1]

# c d liu vo
n, m = map(int, input().split())
a = list(map(int, input().split()))

# Tnh ton v in ra kt qu
result = min_time_to_balance_columns(n, m, a)
print(result)
```

```python
def min_time_to_balance_columns(n, m, a):
    prefix_sum = [0] * (n + 1)
    for i in range(n):
        prefix_sum[i + 1] = prefix_sum[i] + a[i]

    def calc_time(i, j):
        total = prefix_sum[j + 1] - prefix_sum[i]
        mean = total // (j - i + 1)
        time = 0
        for k in range(i, j + 1):
            time += abs(a[k] - mean)
        return time

    dp = [[float('inf')] * (n + 1) for _ in range(m + 1)]
    for i in range(n + 1):
        dp[0][i] = 0

    for k in range(1, m + 1):
        for i in range(k, n + 1):
            for j in range(k - 1, i):
                dp[k][i] = min(dp[k][i], dp[k - 1][j] + calc_time(j, i - 1))

    return dp[m][n]

# c d liu vo
n, m = map(int, input().split())
a = list(map(int, input().split()))

# Tnh ton v in ra kt qu
result = min_time_to_balance_columns(n, m, a)
print(result)
```

```python
def min_time_to_balance_columns(n, m, a):
    prefix_sum = [0] * (n + 1)
    for i in range(n):
        prefix_sum[i + 1] = prefix_sum[i] + a[i]

    def calc_time(i, j):
        total = prefix_sum[j + 1] - prefix_sum[i]
        mean = total // (j - i + 1)
        time = sum(abs(a[k] - mean) for k in range(i, j + 1))
        return time

    dp = [[float('inf')] * (n + 1) for _ in range(m + 1)]
    dp[0][0] = 0

    for k in range(1, m + 1):
        for i in range(k, n + 1):
            for j in range(k - 1, i):
                dp[k][i] = min(dp[k][i], dp[k - 1][j] + calc_time(j, i - 1))

    return dp[m][n]

# c d liu vo
n, m = map(int, input().split())
a = list(map(int, input().split()))

# Tnh ton v in ra kt qu
result = min_time_to_balance_columns(n, m, a)
print(result)
```
